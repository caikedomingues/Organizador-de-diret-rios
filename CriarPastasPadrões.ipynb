```python
import os

import json
```

```python
caminho_extensao = "extensoes.json"

with open(caminho_extensao, "r") as extensao:
    
    dicionario_extensao = json.load(extensao)
    
    
```

```python
for extensao in dicionario_extensao:
    
    os.makedirs(dicionario_extensao[extensao], exist_ok=True)
```
