Bibliotecas Necessárias

```python
# Biblioteca do python que tem como objetivo conectar e manipular o sistema
# de arquivos do sistema operacional.
import os

# Biblioteca que tem como objetivo manipular arquivos json
import json
```

Definindo a pasta que será acessada 

```python
# Ira conter o caminho da pasta de download que contém os arquivos que serão
# organizados.
pasta_download = r"C:\Users\caike\Downloads" 
```

Acessando a lista de arquivos da pasta

```python
# Função da biblioteca os que tem como objetivo listar os arquivos e diretórios
# de uma pasta. Ela ira receber como argumento o caminho da pasta de download.
lista_arquivos = os.listdir(pasta_download)

lista_arquivos
```

carregando o arquivo de extensões: Esse arquivo irá conter um dicionário com as extensões padrões que o sistema localizara. Esse arquivo também irá armazenar as extensoes adicionadas pelo usuário

```python
# Ira conter o caminho do arquivo json que conterá as extensões do sistema
caminho_extensoes = "extensoes.json"

# open: Função nativa do python que tem como objetivo acessar e manipular
# arquivos, vamos usar ela para ler os valores do arquivo de extensões.
# A função recebe como argumento:

# caminho_extensões: Caminho do arquivo json de extensões.

# r: Comando de leitura de arquivos

# extensoes: ira conter os valores do arquivo lido
with open(caminho_extensoes, "r") as extensoes:
    
    # dicionário_extensões: Ira conter o dicionário presente no arquivo json
    
    # load: Função da biblioteca json que irá converter os valores do arquivo
    # em um dicionário.
    
    # Observação: Como normalmente a open retorna os valores como texto
    # (string), é necessário usar a função load que converte os dados
    # em um dicionário.
    dicionario_extensoes = json.load(extensoes)
    

# Ira imprimir os valores do arquivo
print(dicionario_extensoes)
```

Ira adicionar uma nova extensão no script (esse trecho não é obrigatório)

```python
# Ira coletar o nome da extensão enviada pelo usuário.
# lower: Ira converter todas as letras maiusculas em minusculas
nova_extensao = input('Escreva o nome da nova extensão').lower()

# Como os nomes de extensões possuem um "." antes do nome, vamos
# adicionar o "." antes do nome passado pelo usuário.
nova_extensao = "." + nova_extensao

# Ira coletar o nome da pasta que conterá os arquivos da nova extensão
# upper: Por padrão, o sistema irá criar os nomes das pastas em maiusculo.
nova_pasta = input('Escreva o nome da pasta que os arquivos serão salvos').upper()

# Ira adicionar no dicionário a nova extensão e a nova pasta criada
# pelo usuário
dicionario_extensoes[nova_extensao] = nova_pasta

# Ira verificar se o campo de novas extensões está vázio
if nova_pasta != '':
    
    # Se essa condição for verdadeira, vamos criar o diretório da
    # nova extensão
    
    # makedirs: Função da biblioteca os que tem como objetivo criar
    # novos diretórios. 
    
    # exist_ok: quando esse argumento recebe o valor True, ele 
    # verifica se o diretório criado ja existe no sistema, ou
    # seja, se o diretório criado ja existir no sistema, ele irá
    # simplesmente ignorar a criação da pasta.
    os.makedirs(nova_pasta, exist_ok=True)

if nova_extensao == '.' and nova_pasta == '':
    
    dicionario_extensoes.pop('.', '')


with open(caminho_extensoes, "w") as extensoes:
    
    extensoes = json.dump(dicionario_extensoes, extensoes, indent=4)
    



dicionario_extensoes
```

```python
lista_ja_existentes = []


for arquivo in lista_arquivos:
    
    nome, extensao = os.path.splitext(arquivo)
    
    if extensao in dicionario_extensoes:
        
        pasta_destino = dicionario_extensoes[extensao]
                
        arquivos_na_pasta_destino = os.listdir(pasta_destino)
        
        if arquivo in arquivos_na_pasta_destino:
                        
            print(f"O arquivo {arquivo} já existe na pasta {pasta_destino}")
            
            lista_ja_existentes.append(arquivo)
        
        else:
            
            os.rename(f"{pasta_download}/{arquivo}", f"{pasta_destino}/{arquivo}")
            
            print(f"O arquivo {arquivo} foi movido para a pasta {pasta_destino}")
    
    else:
        
        nao_identificados = "NAO IDENTIFICADOS"
        
        os.rename(f"{pasta_download}/{arquivo}", f"{nao_identificados}/{arquivo}")
        
        
    
    
```
