Arquivo de extensões: Esse arquivo irá conter um dicionário com as extensões
padrões que o sistema localizara. Esse arquivo também irá armazenar as extensoes adicionadas pelo usuário

```python
caminho_extensoes = "extensoes.json"

with open(caminho_extensoes, "r") as extensoes:
    
    dicionario_extensoes = json.load(extensoes)
```

Ira adicionar uma nova extensão no script

```python
nova_extensao = input('Escreva o nome da nova extensão')

nova_pasta = input('Escreva o nome da pasta que os arquivos serão salvos').upper()

dicionario_extensoes[nova_extensao] = nova_pasta

dicionario_extensoes.pop('', '')

with open(caminho_extensoes, "w") as extensoes:
    
    extensoes = json.dump(dicionario_extensoes, extensoes, indent=4)


dicionario_extensoes
```
